## Группировки

Инструмент позволяет создать самый простой вид сводных таблиц – объединить данные по полю или нескольким полям и применить к ним одну агрегатную функцию:

К методу `groupby` можно применить следующие агрегатные функции:
* `max()`, `min()` – наибольшее и наименьшее значение
* `sum()`, `count()` – сумма и количество элементов
* `median()`, `mean()` – медиана и среднее
* [другие](https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html)
